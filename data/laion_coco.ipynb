```python
from datasets import load_dataset

dataset = load_dataset("visheratin/laion-coco-nllb")
```

```python
import pandas as pd
```

```python
dataset
```

```python
dataset.set_format(type='pandas')

df_all = dataset['train'][:]
```

```python
df_all = pd.concat([df_all, dataset['test'][:]])
```

```python
df_all
```

```python
df_all.drop(columns=['id', 'score'], inplace=True)
```

```python
df_all['caption_description_ru'] = df_all['captions'].apply(lambda x: [x[i][1] for i in range(len(x)) if x[i][0] == 'rus_Cyrl'])
```

```python
df = df_all[df_all['caption_description_ru'].apply(lambda x: len(x) != 0)].copy()
```

```python
df.loc[:, 'caption_description_ru'] = df.caption_description_ru.apply(lambda x : x[0])
```

```python
df.drop(columns=['captions'], inplace=True)
df.rename(columns={'url' : 'image_url', 'eng_caption' : 'caption_description_en'}, inplace=True)
```

```python
df
```

```python
json = df.to_json(orient="records")
with open("laion_coco.json", "w") as outfile: 
    outfile.write(json)
```
