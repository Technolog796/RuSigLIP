```python
from tensorflow.keras.datasets import cifar10
import pandas as pd
import numpy as np
```

```python
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
```

```python
X_train.shape
```

```python
labels_en = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
labels_ru = ['самолет', 'автомобиль', 'птица', 'кот', 'олень', 'собака', 'лягушка', 'лошадь', 'корабль', 'грузовик']
```

```python
y_en = []
y_ru = []
for i in range(y_train.shape[0]):
  y_en.append(labels_en[y_train[i][0]])
  y_ru.append(labels_ru[y_train[i][0]])
```

```python
new_X = []
for i in range(X_train.shape[0]):
  new_X.append(X_train[i])
```

```python
df = pd.DataFrame({'image': new_X, 'caption_description_en' : y_en, 'caption_description_ru' : y_ru})
```

```python
df.head()
```

```python
df.iloc[0]['image']
```

```python
json = df.to_json(orient="records")
with open("cifar10_en_ru.json", "w") as outfile:
    outfile.write(json)
```
